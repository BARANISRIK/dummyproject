```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
idt=load_iris()
```

# EDA

```python
idt.keys()
```

```python
idt.target_names
```

```python
idt.feature_names
```

```python
dt=pd.DataFrame(idt.data,columns=idt.feature_names)
```

```python
display(dt)
```

```python
y=pd.DataFrame(idt.target)
```

```python
y.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
dt['y']=y
dt['flower_name'] = dt.y.apply(lambda x: idt.target_names[x])
```

```python
dt.head()
```

```python
sns.pairplot(dt,hue="flower_name")
plt.show()
```

```python
dt.columns
```

```python
X=dt.drop(["y","flower_name"],axis=1)
```

```python
X.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
```

```python
knn.fit(X,y)
```

```python
predict=knn.predict([[5.1,3.5,1.4,0.2]])
```

```python
idt.target_names[predict]
```
